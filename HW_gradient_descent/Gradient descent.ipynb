```python
#import necessary library
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ipywidgets import interact
```

## Function
$2x ^ 3 - 3x ^ 2 - 36x + 14$

```python
def f(x):
    return 2 * x ** 3 - 3 * x ** 2 - 36 * x + 14
```

```python
x = np.linspace(0, 5, 100)
y = f(x)
plt.plot(x, y)
```

## Loss function
$6x ^ 2 - 6x - 36$

```python
def df(x):
    return 6 * x ** 2 - 6 * x -36
```

```python
def gradient_descent(df, lr, x_min = 0, x_max = 5, xEpochNum = 100):
    xEpoch = np.zeros(xEpochNum)
    for i in range(xEpochNum):
        if(i == 0):
            xEpoch[i] = np.random.uniform(x_min, x_max, 1)
        else:
            xEpoch[i] = xEpoch[i - 1] - lr * df(xEpoch[i - 1])
    return xEpoch
```

```python
x = np.linspace(0, 5, 100)
gradientX = gradient_descent(df, 0.001)

plt.plot(x, f(x), c = "b")
plt.scatter(gradientX, f(gradientX), c = "r")
plt.show()
```
